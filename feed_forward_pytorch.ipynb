```python
import torch
import torch.nn as nn
import torch.optim as optim

x = torch.tensor([0.5, 0.2, 0.1], dtype=torch.float32)
y = torch.tensor([1.0], dtype=torch.float32)

model = nn.Sequential(
    nn.Linear(3, 2),
    nn.Sigmoid(),
    nn.Linear(2, 1),
    nn.Sigmoid()
)

with torch.no_grad():
    model[0].weight = nn.Parameter(torch.tensor([
        [0.1, 0.2, 0.3],
        [0.4, 0.3, 0.6]
    ]))
    model[0].bias = nn.Parameter(torch.tensor([0.01, 0.02]))

    model[2].weight = nn.Parameter(torch.tensor([[0.2, 0.3]]))
    model[2].bias = nn.Parameter(torch.tensor([0.05]))

output = model(x)
print("Final Output:", output.item())

criterion = nn.MSELoss()
loss = criterion(output, y)
print("Loss:", loss.item())

optimizer = optim.SGD(model.parameters(), lr=0.1)
optimizer.zero_grad()
loss.backward()
optimizer.step()

print("Updated w2:", model[2].weight.data)
```
